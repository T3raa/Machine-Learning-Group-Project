# Exploratory Data Analysis (EDA)
## Student Dropout Prediction

This notebook performs an exploratory data analysis on the cleaned dataset 
`student_cleaned.csv`.

Objective:
- Assess data quality
- Analyze target distribution
- Understand feature behavior
- Identify key drivers of dropout
- Evaluate readiness for machine learning modeling

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("student_cleaned.csv")
TARGET = "Target"

df[TARGET] = df[TARGET].astype(int)

print("Dataset shape:", df.shape)
df.head()
```

```python
df.info()
df.describe()
```

```python
# Missing values
missing = (df.isna().mean() * 100).sort_values(ascending=False)
missing[missing > 0]
# Duplicates
df.duplicated().sum()
# Low variance features
df.var(numeric_only=True).sort_values().head()
```

```python
df[TARGET].value_counts()
df[TARGET].value_counts(normalize=True)
dropout_rate = df[TARGET].mean()
baseline_acc = max(dropout_rate, 1 - dropout_rate)

print("Dropout rate:", round(dropout_rate*100,2), "%")
print("Majority baseline accuracy:", round(baseline_acc*100,2), "%")

sns.countplot(x=TARGET, data=df)
plt.title("Target Distribution (0=Graduate, 1=Dropout)")
plt.show()
```

### Interpretation

This plot shows the distribution of the target variable:

- 0 = Graduate  
- 1 = Dropout  

The dataset is imbalanced.

Approximately 69% of students belong to class 0 (Graduate), while about 31% belong to class 1 (Dropout).

This means that:
- A naïve model predicting only the majority class would already achieve about 69% accuracy.
- Accuracy alone is not a reliable evaluation metric for this problem.
- Metrics such as Recall, Precision, F1-score, and ROC-AUC will be more appropriate.

Implications for modeling:
- Class imbalance must be considered during model training.
- Techniques such as class weighting or resampling may be necessary.

```python
#Numeric Features Identification
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols.remove(TARGET)

numeric_cols
```

```python
#Univariate Analysis (distributions)
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

## Summary of Univariate Analysis

The univariate analysis of the selected numerical and binary variables highlights several important structural patterns in the dataset.

Academic performance variables show distinct distributions:
- *Curricular units 1st sem (grade)* is approximately centered around average passing values, with a small concentration of very low grades that may represent academically at-risk students.
- *Curricular units 1st sem (approved)* displays a positively skewed distribution, with most students approving between 4 and 7 units and a smaller group with zero approvals.

Financial-related variables are clearly imbalanced:
- *Tuition fees up to date* shows that the majority of students are financially regular.
- *Scholarship holder* indicates that only a minority receive financial aid.
- *Debtor* is highly imbalanced, with most students not classified as debtors.

Demographic and status variables reveal structural patterns:
- *Age at enrollment* is strongly right-skewed, with most students enrolling between 18 and 25 years old and a long tail of older students.
- *Displaced* appears relatively balanced between categories.

Overall, several variables show skewness and imbalance, particularly binary financial indicators and age. These patterns suggest that:

- Scaling will be necessary for numerical variables.
- Skewed variables (especially Age at enrollment and Approved units) may benefit from transformation or binning.
- Imbalanced binary variables may carry strong predictive signal despite their lower frequency.

These observations provide a foundation for the next phase: feature engineering and multivariate analysis.

```python
#Skewness Analysis
df[numeric_cols].skew().sort_values(ascending=False)
```

```python
#Outliers Analysis 
def outlier_percentage(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return ((series < lower) | (series > upper)).mean() * 100

for col in numeric_cols:
    print(col, "outlier %:", round(outlier_percentage(df[col]),2))
```

```python
#Correlation Matrix 
corr = df[numeric_cols + [TARGET]].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Matrix")
plt.show()
```

### Interpretation

The correlation matrix shows that the strongest relationships with Target are associated with academic performance and financial variables.

Academic performance variables:
- *Curricular units 1st sem (grade)* and
- *Curricular units 1st sem (approved)*

Both show a moderately strong negative correlation with Target, indicating that better academic performance is associated with a lower probability of dropout.

Financial variables:
- *Tuition fees up to date* shows a negative correlation with Target.
- *Debtor* shows a positive correlation with Target.
- *Scholarship holder* presents a weaker but still relevant association.

Age at enrollment shows a moderate positive correlation with Target, suggesting that older students may have a higher likelihood of dropout.

Regarding multicollinearity:
- There is a strong positive correlation between Grade and Approved units.
- Financial variables such as Tuition fees up to date and Debtor also show noticeable association.

These relationships suggest that academic progression and financial stability are the main structural drivers of the outcome.

```python
#Correlation With Target (Feature Ranking)
corr[TARGET].abs().sort_values(ascending=False)
```

```python
#Multicollinearity Check
corr_abs = corr.abs()
np.fill_diagonal(corr_abs.values, 0)

high_corr_pairs = (
    corr_abs.stack()
    .sort_values(ascending=False)
    .reset_index()
    .rename(columns={"level_0": "Var1", "level_1": "Var2", 0: "AbsCorr"})
)

high_corr_pairs[high_corr_pairs["AbsCorr"] > 0.75].iloc[::2]
```

## Overall Multivariate Insights

The multivariate analysis confirms:

1. Academic performance is the strongest predictor of dropout.
2. Financial instability (Debtor, Tuition status) increases dropout probability.
3. Age has a moderate effect but is secondary to academic variables.
4. Academic variables are strongly correlated (potential multicollinearity).
5. Financial and academic dimensions interact structurally.

Dropout appears to be driven by the combination of:
- Low academic performance
- Financial difficulty
- Higher enrollment age

## Summary of Bivariate Analysis (Feature vs Target)

The bivariate analysis evaluates the relationship between each feature and the Target variable (0 = Graduate, 1 = Dropout).

### Academic Performance Variables

**Curricular units 1st sem (grade)**  
There is a clear separation between the two groups.  
Students who graduated (Target = 0) show significantly higher median grades compared to dropouts (Target = 1).  
The difference in means confirms this gap, indicating that academic performance in the first semester is strongly associated with the final outcome.

**Curricular units 1st sem (approved)**  
A similar pattern emerges.  
Graduated students approved more curricular units on average than dropouts.  
The median and interquartile ranges show noticeable separation between groups.  
This reinforces the importance of early academic progression as a structural determinant of dropout risk.

→ Conclusion: First semester academic performance is one of the strongest predictors of Target.

---

### Financial Variables

**Tuition fees up to date**  
Students who are up to date with tuition payments are predominantly in the Graduate group.  
Dropout students show a higher proportion of unpaid tuition status.  
The difference in group means confirms this financial association.

**Scholarship holder**  
Graduated students have a higher proportion of scholarship holders compared to dropouts.  
Although the separation is weaker than academic variables, financial support appears to act as a protective factor.

**Debtor**  
Dropout students are significantly more likely to be classified as debtors.  
This variable shows strong separation between classes and represents financial instability.

→ Conclusion: Financial stability variables (Tuition status and Debtor) are strongly associated with dropout risk.

---

### Demographic Variable

**Age at enrollment**  
Dropout students tend to be older on average than graduates.  
The distribution for Target = 1 is shifted toward higher ages.  
This suggests that non-traditional students may face higher dropout risk.

---

### Mobility Variable

**Displaced**  
The separation between groups is relatively weak.  
Both categories appear in both Target groups with limited difference in distribution.  
This suggests a weaker predictive contribution compared to academic and financial features.

---

## Overall Insights from Bivariate Analysis

1. Academic performance variables show the strongest separation between classes.
2. Financial instability (Debtor and unpaid tuition) is strongly associated with dropout.
3. Age at enrollment presents a moderate positive association with dropout.
4. Displaced shows limited discriminative power.

These findings indicate that academic progression and financial stability are the primary drivers of dropout risk, while demographic factors play a secondary role.

The strong separation observed in academic and financial variables suggests high predictive relevance and supports their inclusion in the modeling phase.

```python
#Segmentation Analysis 
if "Debtor" in df.columns:
    print("\nDropout rate by Debtor:")
    print(pd.crosstab(df["Debtor"], df[TARGET], normalize="index") * 100)

if "Scholarship holder" in df.columns:
    print("\nDropout rate by Scholarship holder:")
    print(pd.crosstab(df["Scholarship holder"], df[TARGET], normalize="index") * 100)

if "Displaced" in df.columns:
    print("\nDropout rate by Displaced:")
    print(pd.crosstab(df["Displaced"], df[TARGET], normalize="index") * 100)
```

```python
#Early Risk Pattern
risk_conditions = []

if "Curricular units 1st sem (approved)" in df.columns:
    q25 = df["Curricular units 1st sem (approved)"].quantile(0.25)
    risk_conditions.append(df["Curricular units 1st sem (approved)"] <= q25)

if "Debtor" in df.columns:
    risk_conditions.append(df["Debtor"] == 1)

if len(risk_conditions) >= 2:
    high_risk = df[np.logical_and.reduce(risk_conditions)]
    print("High-risk size:", len(high_risk))
    print("High-risk dropout rate:", round(high_risk[TARGET].mean()*100,2), "%")
    print("Overall dropout rate:", round(df[TARGET].mean()*100,2), "%")
```
